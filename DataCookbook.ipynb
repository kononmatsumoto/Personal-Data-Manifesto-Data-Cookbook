# Recipe 1: Merging and Cleaning Multi-Year Data

**What this recipe does:**

This recipe combines and prepares multi-year air quality data for analysis. It merges two datasets into one, converts the 'Date' column into datetime format for easier time-based operations, and sorts the data chronologically. Finally, it previews the cleaned dataset with key columns. This process ensures your data is properly formatted and organized before conducting further analysis or creating visualizations.

This code is preparing and previewing a cleaned dataset for air quality analysis. Here's what each part does:

1.   Merges two datasets (df1 and df2) into a single DataFrame df using pd.concat().

2.   Converts the Date column from string format to datetime format so time-based operations can be performed.

3.   Sorts the rows by date and resets the index to ensure chronological order and a clean index.

4.   Previews the cleaned dataset by showing the first few rows of just the key columns: Date, Daily Mean PM2.5 Concentration, and Daily AQI Value.

This step ensures your data is in the right format and order before doing further analysis or visualization.


```python
# Load and clean your data
# Combine them into one DataFrame
df = pd.concat([df1, df2])

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')

# Sort values by date
df = df.sort_values('Date').reset_index(drop=True)

# Preview cleaned data
df[['Date', 'Daily Mean PM2.5 Concentration', 'Daily AQI Value']]
```

This table shows multiple daily air quality readings taken on the same date — for example, several entries on 2023-01-01 and 2024-12-31. Each row contains:

* The date of observation

* The Daily Mean PM2.5 concentration (in µg/m³), a key pollutant

* The Daily AQI value, a standardized Air Quality Index score

It highlights that for some days, multiple observations were recorded, possibly from different monitoring devices or locations within the same site. This suggests a need to average or analyze by group when aggregating for trends.

# Recipe 2: Monthly Average (Two Methods) + 1 Visualization:

**What this recipe does:**

This recipe calculates the monthly average PM2.5 concentration from daily air quality data using the GroupBy method. It first extracts the month and year from each date and creates a new column for grouping. Then, it uses groupby() and agg('mean') to compute the average PM2.5 level for each month. The result is a summary table showing how air quality varied month-to-month. This method is especially useful for identifying seasonal trends and patterns in pollution levels while avoiding deprecation warnings that come with older aggregation methods.

**Method 1 GroupBy**

This code calculates the monthly average PM2.5 concentration from daily air quality data in a neat manner without warnings.

Here's what it does:
* **df['Month'] = df['Date'].dt.to_period('M')**: Extracts the month and year from each date (e.g., "2023-01") and places it in a new Month column.

* **df.groupby('Month')['Daily Mean PM2.5 Concentration'].agg('mean')**: Groups the data by month and calculates the mean PM2.5 concentration for each month.

* The result is stored in monthly_avg_groupby and printed out, giving a quick summary of how PM2.5 levels varied from month to month.

This is useful for tracking seasonal patterns in air quality and is coded to avoid FutureWarning errors from pandas.

```python
# Method 1: GroupBy using 'mean' string
df['Month'] = df['Date'].dt.to_period('M')
monthly_avg_groupby = df.groupby('Month')['Daily Mean PM2.5 Concentration'].agg('mean')

# Print the result
print("Monthly Average PM2.5 using groupby (no warning):\n")
monthly_avg_groupby
```

**What this recipe does:**

This recipe calculates monthly average PM2.5 concentrations using NumPy’s mean() function while avoiding warnings and producing a clean, labeled output. It begins by setting the Date column as the DataFrame index to enable time-based grouping. The .resample('ME') method groups data by each month’s end. A list comprehension applies np.mean() to each group, and the results are wrapped in a pd.Series() with a custom column name. The index is then converted to datetime format for clarity. This approach produces a streamlined Series of monthly averages, ideal for clean reporting and visualization.

**Method 2 NumPy**

This code calculates monthly average PM2.5 concentrations using NumPy's mean() function without triggering any warnings and cleans up the output format.

What each part does:
* **df.set_index('Date')**: Sets the Date column as the index so you can use time-based operations like .resample().

* **.resample('ME')**: Groups the data by month-end ('ME'), creating one group per month.

* **[np.mean(group) for _, group in resampled]**: Loops through each monthly group and calculates the average PM2.5 using np.mean().

* **pd.Series(..., name='Monthly Avg PM2.5')**: Wraps the results in a pandas Series with a proper name, avoiding the default header 0.

* **index=pd.to_datetime(list(resampled.groups.keys()))**: Sets the index of the Series to the month-end dates of each group.

* **print(...) and monthly_avg_resample**: Outputs a clean, labeled Series of monthly averages.

```python
# Resample using NumPy mean, then assign a proper column name
df_resample = df.set_index('Date')
resampled = df_resample['Daily Mean PM2.5 Concentration'].resample('ME')
monthly_avg_resample = pd.Series(
    [np.mean(group) for _, group in resampled],
    index=pd.to_datetime(list(resampled.groups.keys())),
    name='Monthly Avg PM2.5'
)

print("\nMonthly Average PM2.5 using resample with NumPy (clean header):\n")
monthly_avg_resample
```

This table displays the monthly average PM2.5 concentrations from January 2023 through December 2024. It summarizes daily pollution readings into a single mean value for each month, helping reveal broader air quality trends.

You can clearly see seasonal patterns — for example:

* August 2023 shows a significant spike in PM2.5 levels (15.58 µg/m³), likely due to wildfire smoke or other environmental events.

* The values in spring 2024 are notably lower, indicating cleaner air during that period.

This kind of monthly aggregation is useful for comparing pollution levels across years and identifying high-risk periods.

# Recipe 3: Loop Through Years and Compare Monthly Averages

**What this recipe does:**

This recipe compares monthly average PM2.5 levels across years using a multi-line chart. It groups data by year and month, calculates monthly averages, and plots each year as a separate line to show trends and seasonal differences.

This illustrates how to create a multi-line plot to compare monthly average PM2.5 concentrations by year. This is useful for seeing seasonality and year-to-year differences.

* **df['Year'] = df['Date'].dt.year**: Extracts the year from the 'Date' column and stores it in a new 'Year' column.

* **df['Month_Num'] = df['Date'].dt.month**: Extracts the month number (1–12) from the 'Date' column and stores it in 'Month_Num'.

* **groupby(['Year', 'Month_Num']).mean().unstack(0)**: Groups the data by year and month, calculates the mean PM2.5 per group, and transposes the result so that each year becomes a column.

* **plt.figure(figsize=(12, 6))**: Creates a 12x6 inch figure for the plot.

* **for year in monthly_by_year.columns:.**: Loops through each year (now a column) and plots its monthly values on the same graph.

* **plt.plot(.)**: Plots the monthly averages for a single year as a line.

* **plt.title("Monthly PM2.5 Comparison by Year"**): Adds a title to the chart.

* p**lt.xlabel("Month") / plt.ylabel("PM2.5 (µg/m³)")**: Adds labels to the axes for clearer results.

* **plt.xticks(range(1, 13))**: Sets the x-axis ticks to show months 1 to 12.

* **plt.legend(title="Year")**: Adds a legend title to specify which line is for which year.

* **plt.grid(True)**: Adds gridlines to allow easier comparison by sight.

* **plt.show()**: Plots the final figure.

This code calculates and plots the average monthly PM2.5 levels for each year in the data. It first extracts the year and month from the 'Date' column and puts them into new columns called 'Year' and 'Month_Num'. It then groups the data by year and month number, calculates the mean PM2.5 level for each group, and uses.unstack(0) to pivot the years into columns. In the plotting section, a graph is plotted, and for each year, a separate line is plotted showing the trend of the PM2.5 level month by month (1 to 12). In the x-axis, "Month" is placed and on the y-axis the PM2.5 concentration is shown in micrograms per cubic meter (µg/m³). A legend is added to distinguish the years, grid lines improve readability, and finally plt.show() plots the whole multi-line chart of PM2.5 levels across different years.


```python
# Recipe 3: Loop Through Years and Compare Monthly Averages

# Extract year and month from the date
df['Year'] = df['Date'].dt.year
df['Month_Num'] = df['Date'].dt.month

# Group by year and month, calculate mean PM2.5, and restructure for plotting
monthly_by_year = df.groupby(['Year', 'Month_Num'])['Daily Mean PM2.5 Concentration'].mean().unstack(0)

# Plot comparison of PM2.5 levels by year
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

for year in monthly_by_year.columns:
    plt.plot(monthly_by_year.index, monthly_by_year[year], label=str(year))

plt.title("Monthly PM2.5 Comparison by Year")
plt.xlabel("Month")
plt.ylabel("PM2.5 (µg/m³)")
plt.xticks(range(1, 13))
plt.legend(title="Year")
plt.grid(True)
plt.show()
```

This is a month-to-month comparison of average PM2.5 concentrations for the years 2023 and 2024. One line for each year is employed to show how particulate matter concentrations (PM2.5 in µg/m³) varied over the 12 calendar months.

From the chart, we can see that PM2.5 levels trended quite similarly seasonally for both years, starting high in winter and decreasing in the spring. But there's a giant spike in August of 2023, with readings spiking sharply compared to the same month of 2024. This may suggest a likely pollution episode—like smoke from fires or factory emissions—specific to August of 2023.

Conversely, 2024 shows a less volatile curve with lower overall PM2.5 concentrations during the year, particularly during the summer. This could indicate improved air quality or more moderate environmental conditions that year.

Overall, the visualization helps to determine seasonal air pollution trends, year-to-year differences, and outliers worthy of further analysis.

# Recipe 4: Flagging and Counting High AQI Days

**What this recipe does:**

This code identifies and counts days where the Air Quality Index (AQI) exceeds 100, indicating unhealthy air. It flags these days with a Boolean column (True if AQI > 100), then sums the True values to count the total number of high-AQI days.


This code calculates and adds the number of days with unhealthy air quality, as per the Air Quality Index (AQI).

* **df['High AQI'] = df['Daily AQI Value'] > 100**: Creates a new Boolean column 'High AQI' which marks each row True if the AQI value is greater than 100 (unhealthy air), and False otherwise.

* **high_aqi_days = df['High AQI'].sum()**: Counts the number of True values in column 'High AQI'. Since True is the same as 1 in Python, adding up the column gives the number of days when AQI > 100.

* **print(.)**: Prints the number of high AQI days in a readable sentence:
"Number of days with AQI > 100: 193"

```python
# Recipe 4: Flagging and Counting High AQI Days
df['High AQI'] = df['Daily AQI Value'] > 100
high_aqi_days = df['High AQI'].sum()
print(f"Number of days with AQI > 100: {high_aqi_days}")


```

The result — "Number of days with AQI > 100: 193" — means that out of all the dates in the data set, there were 193 days on which the Air Quality Index (AQI) was greater than 100.

What AQI > 100 means:

* AQI > 100 is unhealthy for sensitive groups such as children, elderly, and
people with respiratory ailments. The elevated values can reflect hazardous air quality.


Interpretation:
* This count highlights how often the air quality was potentially unhealthy throughout the range covered by your data. To have 193 high-AQI days signals a high number of pollution events, which is a concern for environmental and public health monitoring.

High AQI Days per Year (Bar Chart)
This code calculates the number of days with high air pollution (AQI > 100) for each year and visualizes the results as a bar chart. This helps us understand how the frequency of poor air quality days changes annually.

What each part does:

* **.groupby(df['Date'].dt.year).size()**: Groups these high AQI days by year and counts the number of days in each year.

* **plt.figure(figsize=(8, 5))**: Initializes the plot with a figure size of 8 inches by 5 inches.

* **plt.bar(..., color='crimson')**: Plots a bar chart with crimson-colored bars. The x-axis shows years and the y-axis shows the number of high AQI days.

* **plt.title(...) / plt.xlabel(...) / plt.ylabel(...)**: Adds a descriptive title and labels to the x and y axes.

* **plt.grid(axis='y', linestyle='--', alpha=0.7)**: Adds horizontal grid lines to improve readability.

* **plt.tight_layout() / plt.show()**: Ensures layout spacing is clean and displays the final plot.

```python
# Step 2: Count high AQI days per year
high_aqi_by_year = df[df['High AQI']].groupby(df['Date'].dt.year).size()

# Step 3: Bar chart visualization
plt.figure(figsize=(8, 5))
plt.bar(high_aqi_by_year.index.astype(str), high_aqi_by_year.values, color='crimson')
plt.title("Number of High AQI Days (>100) per Year")
plt.xlabel("Year")
plt.ylabel("Number of Days")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

This bar chart shows the number of days during a year when the Air Quality Index (AQI) exceeded 100—a value which means unhealthy air. The x-axis is years (2023 and 2024), and the y-axis is the number of days with a high AQI.

From the chart, we can readily see that 2023 experienced significantly more days with AQI > 100—about 157 days—than almost 36 days in 2024. This suggests a vast improvement in air quality in 2024 or perhaps even a change in contributing factors such as pollution sources, weather conditions, or reporting coverage. The red-hued bars also help to highlight this stark contrast visually.

# Recipe 5: How to Filter DataFrames (Two Ways)

**What this recipe does:**

This recipe demonstrates two ways to filter a DataFrame to identify days in Washington state when the Air Quality Index (AQI) exceeded 100. It uses Boolean indexing (Method A) and the .query() method (Method B) to isolate rows based on multiple conditions, then displays selected columns (Date, County, AQI Value). This helps highlight unhealthy air quality events by location and date.

This code demonstrates how to filter a DataFrame using Boolean indexing to isolate specific rows of interest. In this case, it identifies days in Washington state when the Air Quality Index (AQI) was greater than 100.

Here's what each part does:
* **filter_a = df[(df['State'] == 'Washington') & (df['Daily AQI Value'] > 100)]**:
This line filters the DataFrame df to include only rows where the state is "Washington" and the AQI exceeds 100. The & operator combines both conditions, and the result is stored in filter_a.

* **print(...)**: Prints a label describing what the output shows: the first 5 rows that match the filter.

* **filter_a[['Date', 'County', 'Daily AQI Value']]**: Displays only selected columns (Date, County, and Daily AQI Value) from the filtered data.

Result:
The output shows 193 rows where high AQI levels (above 100) were recorded in various counties in Washington, with the first few entries shown as a sample. This is useful for identifying pollution hotspots within a specific state.

```python
# Recipe 5: How to Filter DataFrames (Two Ways)
# Method A: Boolean indexing
filter_a = df[(df['State'] == 'Washington') & (df['Daily AQI Value'] > 100)]

# View filtered results
print("\nFiltered rows (first 5) using Boolean indexing:")
filter_a[['Date', 'County', 'Daily AQI Value']]

```

This output is a tabular preview (not a graph) showing a subset of filtered data where daily AQI (Air Quality Index) values exceeded 100 in Washington state. It displays the first and last few rows of the filtered results to give a sense of the data distribution across time and counties.

What the table shows:
Columns:

* Date: The specific date the high AQI was recorded.

* County: The county in Washington where the AQI exceeded 100.

* Daily AQI Value: The measured AQI for that day — all values are above the unhealthy threshold of 100.

Summary:

* The preview includes counties like Kittitas, Asotin, and Pierce with AQI values ranging from 102 to as high as 167.

* The full filtered dataset contains 193 rows, indicating there were 193 high-AQI days recorded across multiple counties in Washington over the analyzed period.

    * The 'Daily AQI Value' is more than 100
